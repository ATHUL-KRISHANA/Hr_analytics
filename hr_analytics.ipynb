**Feature Selection**

```python
#use chi2
feature=df.drop(['is_promoted','no_of_trainings','age','previous_year_rating','length_of_service','avg_training_score','employee_id'],axis=1)
```

```python
from scipy.stats import chi2_contingency
score=[]
for i in feature.columns:
  ct=pd.crosstab(df['is_promoted'],feature[i],margins=True)
  res=chi2_contingency(ct)[1]
  score.append(res)
```

```python
a=pd.DataFrame()
a['column_name']=feature.columns
a['value']=score
a
```

```python
sns.barplot(data=a,x='column_name',y='value')
plt.xticks(rotation=90)
```

```python
for i in a.index:
  if a.loc[i,'value']>0.05:
    print(a.loc[i,'column_name'])
```

```python
df.drop(['employee_id','gender'],axis=1,inplace=True)
df
```

Feature Ecoding

```python
#ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
o=OrdinalEncoder(categories=[['Below Secondary',"Bachelor's","Master's & above"]])
df[['education']]=o.fit_transform(df[['education']])
```

```python
#use get_dummies encoding technique
demo=pd.get_dummies(df[['department'	,'region'		,'recruitment_channel']],drop_first=True,dtype=int)
```

```python
demo.shape
```

```python
df_final=pd.concat([df,demo],axis=1)
df_final
```

```python
#final dataset
df_final.drop(['department'	,'region'		,'recruitment_channel'],axis=1,inplace=True)
df_final
```

```python
x=df_final.drop(['is_promoted'],axis=1).values
x
```

```python
y=df_final['is_promoted'].values
y
```

**Model Building**

```python
#splitting the model into train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
#standard scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#perform eachh algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
rf=RandomForestClassifier()
dt=DecisionTreeClassifier()
be=BernoulliNB()
s=SVC()
knn=KNeighborsClassifier()
objects=[knn, be, s, dt, rf]
for i in objects:
  i.fit(x_train,y_train)
  result=i.predict(x_test)
  print('result is :',i, result)
  print('accuracy score of ',i, accuracy_score(result, y_test))
  print('classification report ', classification_report(result, y_test))
  print('*'*100)
```

**Hyperparameter Tunning**

```python
from sklearn.model_selection import RandomizedSearchCV
param_grid = {'n_estimators': [50, 100, 200],'max_depth': [None, 5, 10],'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4],'bootstrap': [True, False]}
```

```python
rf1=RandomForestClassifier()

rf_grid = RandomizedSearchCV(estimator = rf1,param_distributions = param_grid,cv = 5,verbose=2,n_jobs = -1)

```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf2=RandomForestClassifier(n_estimators= 200,min_samples_split= 10, min_samples_leaf= 2,max_depth= None,bootstrap= False)
rf2.fit(x_train,y_train)
res=rf2.predict(x_test)
print(accuracy_score(res,y_test))
```
